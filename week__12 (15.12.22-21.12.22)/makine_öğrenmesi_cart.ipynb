###**Regresyon Problemleri için Cost/Loss/Objective Fonsksiyonu**

* Lineer denklemlerde doğruyu nereye koyacağız problemi varken, karar ağaçlarında da dalı nereden böleceğiz problemi vardır. Bölme noktaları değiştikçe hata değişmektedir. 

* Yukarıdaki ağacı oluşturmak için SSE'nin minimum olduğu noktalara göre bölgelenmeler/yapraklanmalar/kutulanmalar oluşturulur.  

<u>**Hiperparametreler**</u>

* **max_depth :** Karar Ağacının maksimum derinliğini ifade eder. Değer girilmezse limitsiz olur. Model overfit(aşırı uyum) olmuşsa düşürülmesi gerekir.


* **min_samples_split :** Bir düğümün bölünmeden önce sahip olması gereken minimum örnek sayısıdır.

* **min_samples_leaf :** Bir yaprağın sahip olması gereken minimum örnek sayısıdır.

* **min_weight_fraction :** min_samples_leaf’e benzer.Ağırlıklı örneklerin, toplam örnekler içerisindeki oranı. Ağacın dengeli gitmesi için kullanılır.

* **max_leaf_nodes :** Maksimum yaprak sayısı.

* **max_features :** En iyi bölünmeyi ararken göz önünde bulundurulması gereken featureların sayısı.

* Genellenebilirliği kaybedip, aşırı öğrenmeye gider. Model overfit (aşırı uyum) olmuşsa genellikle önce **max_depth** hiperparametresi düşürülür.

* Sınıflandırma problemi için **gini katsayısı** ya da **entropi** kullanılabilir. Böylelikle tahmin başarısına yönelik bilgi elde edilir. Ne kadar düşük olursa o kadar iyidir. Sınıflandırma problemleri için yapılan hatanın değerlendirilmesi sağlanır.

* **Entropi,** çeşitlilik demektir. Ne kadar düşük o kadar iyidir. Tahmin değerleri elde edileceğinden bunların çeşitli olmamasını bekleriz. $\hat{p}_{mk}$ değeri 1 sınıfının gerçekleşmesi olasılığıdır. 

* **Gini,** alt kümenin saflık değeridir. Gini değeri 0 ile 1 arasında bir sonuç alır ve sonuç 0’a ne kadar yakınsa o kadar iyi ayrım yapmış olur.

* Gini ve Entropi arasında çok büyük bir fark yoktur. Entropi daha dengeli bir ağaç çıkarmaya meyilli iken Gini frekansı fazla olan sınıfı ayrıştırmaya meyillidir.
