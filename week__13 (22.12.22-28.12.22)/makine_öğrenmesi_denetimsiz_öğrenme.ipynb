### ✨ **1. K-Ortalamalar (K-Means)**

* Amaç, gözlemleri birbirlerine olan benzerliklerine göre kümelere ayırmaktır.

* Bazı gözlemler birbirlerine olan yakınlıklarına göre çeşitli uzaklık metrikleri kullanılarak kümelere ayırılır.

**NOT:** Sınıflandırma problemlerinde label (etiket) belli olup sınıflandırılırken,  K-Means'te label (etiket) olmadığı için gözlem birimlerinin kendi içinde birbirlerine benzerliklerine göre gruplara ayrılır.

**Adım 1 :** Küme sayısı belirlenir.

**Adım 2 :** Rastgele k merkez seçilir.

**Adım 3 :** Her bir gözlem için k merkezlere uzaklık hesaplanır.

**Adım 4 :** Her gözlem en yakın olduğu merkeze yani kümeye atanır.

**Adım 5 :** Atama işlemlerinden sonra oluşan kümeler için tekrar merkez hesaplamaları yapılır.

**Adım 6 :** Bu işlem belirlenen bir iterasyon adedince tekrar edilir ve küme içi hata kareler toplamlarının toplamının (total within-cluster variation SSE1+SSE2+SSE3) minimum olduğu durumdaki gözlemlerin kümelenme yapısı nihai kümelenme olarak seçilir. SSE-SSD-SSR'ın düşük olduğu yer merkez olarak belirlenir.

* **Dezavantaj :** Küme sayısının nasıl belirleneceğidir. Bunun için **Elbow yöntemi** kullanılır. Küme sayısı arttıkça hatanın düşmesi beklenir fakat çok fazla kümeye ayırmakta iyi değildir. Bunun için iş bilgisi önem arz etmektedir. Elbow yöntemi sadece fikir verir.
