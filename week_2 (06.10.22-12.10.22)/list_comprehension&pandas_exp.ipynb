### **List Comprehension**

$\star$ Birden fazla satırda yapacağın işlemleri tek satırda yapmanı sağlar.

$\rhd$ liste = ["a","e","ı","i","u","ü", "o","ö"] elemanları list comprehension ile yazdır. 

* tek bir döngü ile comprehension --> [i for i in liste] 

$\rhd$ List Comprehension yapısı kullanarak liste = ["a","e","ı","i","u","ü", "o", "ö"] verisindeki a elemanı hariç diğer elemanları büyük harfe çeviriniz.

* Tek koşul ile --> [i.upper() for i in liste if i!="a"]

$\rhd$ List Comprehension yapısı kullanarak liste = ["a","e","ı","i","u","ü", "o", "ö"] verisindeki a elemanı hariç diğer elemanları büyük harfe çeviriniz ve a harfi olduğu gibi kalsın.

* İki koşul ile --> [i.upper()if i!="a" else i for i in liste ]

**Import "car_crashes" dataset**

```python
df = sns.load_dataset("car_crashes")
print(df.head())
df.info()
```

#### **GÖREV 1:** List Comprehension yapısı kullanarak car_crashes verisindeki numeric değişkenlerin isimlerini büyük harfe çeviriniz ve başına NUM ekleyiniz.

Not : Numeric olmayan değişkenlerin de isimleri büyümeli. Tek bir list comp. yapısı kullanılmalıdır.

```python
["NUM_"+col.upper() if df[col].dtype!= "O" else col.upper() for col in df.columns]
```

#### **GÖREV 2:** List Comprehension yapısı kullanarak car_crashes verisindeki isminde "no" barındırmayan değişkenlerin isimlerininin sonuna "FLAG" yazınız.

Not: Tüm değişkenlerin isimleri büyük harf olmalı. Tek bir list comp. yapısı kullanılmalıdır.

```python
[col.upper() +"_FLAG" if "no" not in col else col.upper() for col in df.columns]
```

#### **Görev 3:** List Comprehension yapısı kullanarak aşağıda verilen değişken isimlerinden FARKLI olan değişkenlerin isimlerini seçiniz ve yeni bir dataframe oluşturunuz.

og_list = ["abbrev", "no_previous"]

* Önce yukarıdaki listeye göre list comprehension kullanarak new_cols adında yeni liste oluşturunuz.
* Sonra df[new_cols] ile bu değişkenleri seçerek yeni bir df oluşturunuz adını new_df olarak isimlendiriniz.


```python
og_list = ["abbrev", "no_previous"]
new_cols = [col for col in df.columns if col not in og_list]
new_df = df[new_cols]
new_df.head()
```

#### **TASK 15** : survived değişkeninin pclass ve cinsiyet değişkenleri kırılımında sum, count, mean değerlerini bulunuz.

```python
df = sns.load_dataset("titanic")
```
