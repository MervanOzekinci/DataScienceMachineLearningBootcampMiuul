### **Bayesian Average Rating Score**

```python
# Sorting Products with 5 Star Rated
# Sorting Products According to Distribution of 5 Star Rating

# rating ile ilgili bir ortalama değer yani puan dağılımlarının üzerinden olasılıksal bir ortalama hesabı yapar.
# Dilersek bu değerlere göre sıralama yapılabilir, fakat beraberinden bazı sorunlara sebep olacaktır.
# Ya da hibrit bir yaklaşım geliştirilebilir.

def bayesian_average_rating(n, confidence=0.95): #n değeri 1 yıldızdan kaç adet var 2 yıldız kaç adet bunları döndürür
    if sum(n) == 0:
        return 0
    K = len(n)
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    N = sum(n)
    first_part = 0.0
    second_part = 0.0
    for k, n_k in enumerate(n):
        first_part += (k + 1) * (n[k] + 1) / (N + K)
        second_part += (k + 1) * (k + 1) * (n[k] + 1) / (N + K)
    score = first_part - z * math.sqrt((second_part - first_part * first_part) / (N + K + 1))
    return score

df.head()
```

```python
df["bar_score"] = df.apply(lambda x: bayesian_average_rating(x[["1_point",
                                                                "2_point",
                                                                "3_point",
                                                                "4_point",
                                                                "5_point"]]), axis=1)
```

### **Hybrid Sorting: BAR Score + Diğer Faktörler**

 **Rating Products**

 - Average

 - Time-Based Weighted Average

 - User-Based Weighted Average

 - Weighted Rating

 - Bayesian Average Rating Score (Bu yöntem puanları kırpmaktadır. Score biraz aşağı çeker.)


 **Sorting Products**

 - Sorting by Rating

 - Sorting by Comment Count or Purchase Count

 - Sorting by Rating, Comment and Purchase

 - Sorting by Bayesian Average Rating Score Sorting Products with 5 Star Rated)

 - Hybrid Sorting: BAR Score + Diğer Faktörler(BAR Score güveniyoruz fakat tek başına yeterli değil dolayısıyla diğer faktörleride işin içine katmamız gerekli).

```python
def hybrid_sorting_score(dataframe, bar_w=60, wss_w=40):
    bar_score = dataframe.apply(lambda x: bayesian_average_rating(x[["1_point",
                                                                     "2_point",
                                                                     "3_point",
                                                                     "4_point",
                                                                     "5_point"]]), axis=1)
    wss_score = weighted_sorting_score(dataframe)

    return bar_score*bar_w/100 + wss_score*wss_w/100


df["hybrid_sorting_score"] = hybrid_sorting_score(df)

df.sort_values("hybrid_sorting_score", ascending=False).head(20)
```

## IMDB Weighted Rating (2015 yılına kadar geliştirilmiş yaklaşım).

* **weighted_rating** = $\frac{v}{v+M}r + \frac{M}{v+M}C$

* **r** = vote average (filmin ortalama puanı)

* **v** = vote count (bir filme kaç kişi oy vermiş)

* **M** = minimum votes required to be listed in the Top 250 (sıralamaya girmek için gerekli olan minimum **M** değeri )

* **C** = the mean vote across the whole report (currently 7.0) (kitlenin tüm filmlerin genel ortalaması)

C ve M önemli. 

```python
# Film 1:
# r = 8
# M = 500
# v = 1000

# (1000 / (1000+500))*8 = 5.33


# Film 2: Puan aynı, gereken oy sayısı da aynı daha fazla oy alsın.
# r = 8
# M = 500
# v = 3000

# (3000 / (3000+500))*8 = 6.85 arttı

# (1000 / (1000+500))*9.5 = 6.33 puan az olsa da daha çok tercih edilme daha fazla oya sahip olma etkisi

# Film 1:
# r = 8
# M = 500
# v = 1000

# Birinci bölüm:
# (1000 / (1000+500))*8 = 5.33

# İkinci bölüm:
# 500/(1000+500) * 7 = 2.33

# Toplam = 5.33 + 2.33 = 7.66


# Film 2:
# r = 8
# M = 500
# v = 3000

# Birinci bölüm:
# (3000 / (3000+500))*8 = 6.85

# İkinci bölüm:
# 500/(3000+500) * 7 = 1

# Toplam = 7.85

```

```python
M = 2500
C = df['vote_average'].mean()

def weighted_rating(r, v, M, C):
    return (v / (v + M) * r) + (M / (v + M) * C)
```

```python
# Bayesian Average Rating Score

#ilk 5
# 12481                                    The Dark Knight
# 314                             The Shawshank Redemption
# 2843                                          Fight Club
# 15480                                          Inception
# 292                                         Pulp Fiction
```

```python
def bayesian_average_rating(n, confidence=0.95):
    if sum(n) == 0:
        return 0
    K = len(n)
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    N = sum(n)
    first_part = 0.0
    second_part = 0.0
    for k, n_k in enumerate(n):
        first_part += (k + 1) * (n[k] + 1) / (N + K)
        second_part += (k + 1) * (k + 1) * (n[k] + 1) / (N + K)
    score = first_part - z * math.sqrt((second_part - first_part * first_part) / (N + K + 1))
    return score

bayesian_average_rating([34733, 4355, 4704, 6561, 13515, 26183, 87368, 273082, 600260, 1295351])
```
