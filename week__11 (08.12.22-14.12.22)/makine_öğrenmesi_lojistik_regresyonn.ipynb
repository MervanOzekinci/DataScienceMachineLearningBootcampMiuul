 ### **Lojistik Regresyon için Gradient Descent**


Gradient Descent (Gradyan Azalma), rastgele alınan değişkenlerle başlayarak global minimum değerine ulaşmayı amaçlayan en popüler optimizasyon algoritmalarından biridir. Lojistik Regresyon kapsamında, Gradient Descent ile iterasyonlarda kullanılmak üzere Cost Fonksiyonu üzerinden elde edilen J(θ) değerlerine ulaşılır:


Ekran Görüntüsü (1539).png

* $y^{(i)}$ gerçek değerleri, $h_θ(x^{(i)})$ ise bağımlı değişken 1 sınıfının tahminlerinin gerçekleşmesi olasılığını ifade etmektedir.

* Bu problem özelinde Cost Fonksiyonu (Loss Fonksiyon) Log Loss ifadesidir. Her bir gözlem birimi tek tek gezilerek J(θ) hesaplanır.

* Log loss fonksiyonu içerisindeki $h_θ(x^{(i)})$ ifadesi ise bağımlı değişkenin 1 sınıfının gerçekleşmesi olasılığıdır. Dolayısıyla $-y^{(i)}log(h_θ(x^{(i)}))$ ifadesi içerisinde değişkenin 1 sınıfının tahmin edilen gerçekleşme olasılığı ifade eder.

* $(1-y^{(i)})log(1-h_θ(x^{(i)}))$ ifadesi içerisindeki $1-h_θ(x^{(i)})$ kısmında 1 sınıfının tahmin edilen gerçekleşmemesi (yani 0 sınıfının tahmin edilen gerçekleşmesi) olasılığı gözlenir.

* Yukarıdaki formül Corss-Entropi formülüdür. Entropi, çeşitlilik bilgi demektir.

* Entropi ne kadar yüksekse çeşitlilik o kadar fazladır. Varyans vardır.

* Lojistik regresyonda bu çeşitlilik ve bilgi durumu, yapılan tahminleme ile gerçek değer arasındaki durumu değerlendirmek için kullanılır.

* Gerçek değerler ile tahmin edilen değerler açısından entropinin (çeşitliliğin) düşük olmasını bekleriz. 

**Özetle,**
* Gerçek değerler ile o gerçek değerlerin gerçekleşmesi olasılığı ifadeleri birbirine ne kadar yakın ise Loss Değeri (kayıp değeri) o kadar küçük olacaktır.

### **Sınıflandırma Problemlerinde Başarı Değerlendirme** 

⚒ **Karmaşıklık Matrisi ( Confusion Matrix )**

Model kurulduktan sonra ortaya çıkan tahmin edilen değerler sonucunda,
tahmin edilen değerin ve gerçek değerin eşleştiği gözlem birimleri:

* 1 sınıfı için True Positive (TP)
* 0 sınıfı için True Negative (TN)

Tahmin edilen değerin ve Gerçek değerin eşleşmediği gözlem birimleri:

* 1 sınıfı için False Positive (FP) (İstatistikte 1.tip hata olarak betimlenir)
* 0 sınıfı için False Negative (FN) (İstatistikte 2.tip hata olarak betimlenir)

olarak ifade edilir. Yani doğru yapılan işlemlere True ve Yanlış işlemlere ise False ifadesi kullanılır.

Bu durumda eşik değerini belirlemek için ROC Curve (Receiver Operating Characteristic Curve) başarı değerlendirme yöntemi kullanılır. 

⚒ **ROC Curve (Receiver Operating Characteristic Curve)**

$→$ Olası eşik değer değişimlerine karşılık başarı değişimleri nedir ? soruna cevaben kullanılmaktadır. 

Ekran Görüntüsü (1544).png

* Belirlenen thresholdlara göre oluşturulan karmaşıklık matrisinden elde edilen TP rate ve FP rate değerlerine bakarak karşılık gelen noktalar tabloda işaretlenir.

* Bir eğri ne kadar fazla yer kaplıyorsa, model o kadar başarılıdır denir.


⚒ **Area Under Curve (AUC)**

* ROC eğrisi altında kalan alanın integrali alınırsa, Area Under Curve (AUC) elde edilir. 

* ROC eğrisinin tek bir sayısal değer ile ifade edilişidir.

* ROC eğrisi altında kalan alandır. 

* AUC, tüm olası sınıflandırma eşikleri için toplu bir performans ölçüsüdür.
